 Transformation nos données quantitatives en données qualitatives



```python
def categorical(df, column):
    df = df.dropna()
    liste_ = list(df[column].value_counts().index)
    df[column] = df[column].apply(lambda x: liste_.index(x))
    return df
```

```python
listQuant = ["property_type", "room_type", "bed_type"]

for col in listQuant:
    data_clean = categorical(data_clean, col)
print(data_clean)
```

```python
data_clean.info()
```

```python
data_clean['log_price'] = data_clean['log_price'].astype(int)
data_clean['bathrooms'] = data_clean['bathrooms'].astype(int)
data_clean['review_scores_rating'] = data_clean['review_scores_rating'].astype(int)
data_clean['bedrooms'] = data_clean['bedrooms'].astype(int)
data_clean['beds'] = data_clean['beds'].astype(int)
data_clean['cancellation_policy'] = pd.Categorical(data_clean['cancellation_policy'])
```
